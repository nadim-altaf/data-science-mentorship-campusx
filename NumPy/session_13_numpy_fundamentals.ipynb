### What is numpy?

NumPy is the fundamental package for scientific computing in Python. It is a Python library that provides a multidimensional array object, various derived objects (such as masked arrays and matrices), and an assortment of routines for fast operations on arrays, including mathematical, logical, shape manipulation, sorting, selecting, I/O, discrete Fourier transforms, basic linear algebra, basic statistical operations, random simulation and much more.


At the core of the NumPy package, is the ndarray object. This encapsulates n-dimensional arrays of homogeneous data types

### Numpy Arrays Vs Python Sequences

- NumPy arrays have a fixed size at creation, unlike Python lists (which can grow dynamically). Changing the size of an ndarray will create a new array and delete the original.

- The elements in a NumPy array are all required to be of the same data type, and thus will be the same size in memory.

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

- A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays.

#### NumPy arrays are stored and processed using optimized C code, which makes them much faster than Python lists.

### Indexing and Slicing (NumPy)
→ methods used to access single or multiple elements from an array

### Indexing
→ accessing individual element using its position (index)

- **1D indexing** → access element using single index → `a[0]`
- **2D indexing** → access element using row and column index → `a[1,2]`
- **Negative indexing** → access elements from end of array → `a[-1]`

### Slicing
→ extracting a subset of array using range of indices

- **1D slicing** → select elements using start:end → `a[1:4]`
- **Step slicing** → select elements with step size → `a[0:5:2]`
- **2D slicing (rows)** → select rows → `a[0:2, :]`
- **2D slicing (columns)** → select columns → `a[:, 1:3]`
- **Row and column slicing** → select submatrix → `a[0:2, 1:3]`

### NumPy Slicing (Common Patterns)

`a[1:4]`   → elements from index 1 to 3  
`a[:3]`    → first three elements  
`a[::2]`   → every second element  
`a[:,1]`   → full column (2D)  

`a[2:]`    → from index 2 to end  
`a[:-1]`   → all except last element  
`a[-3:]`   → last three elements  
`a[::-1]`  → reverse array  

`a[1:5:2]` → elements from index 1 to 4 with step 2  
`a[:]`     → entire array  
`a[::]`    → entire array (same)  

`a[0,:]`   → first row (2D)  
`a[1,:]`   → second row (2D)  
`a[:,:]`   → full 2D array  
`a[0:2,:]` → first two rows  
`a[:,0:2]` → first two columns  

`a[1:3,1:3]` → middle sub-matrix  
`a[:,-1]`    → last column  
`a[-1,:]`    → last row

### NumPy Iterating

**Definition:**  
Iterating means accessing array elements one by one.

`for x in a`  
→ iterate through 1D array elements  

`for row in a`  
→ iterate row by row (2D array)  

`for row in a:`  
`    for element in row`  
→ iterate each element in 2D array  

`np.nditer(a)`  
→ iterate all elements (any dimension)  

`np.ndenumerate(a)`  
→ iterate with index and value  

`a.flatten()`  
→ convert to 1D then iterate easily  

**Notes**
- Used to access elements one by one  
- `nditer()` best for multi-dimensional arrays  
- `ndenumerate()` gives index + value  
- Vectorized operations are faster than loops

### Stacking

### NumPy Stacking

**Definition:**  
Stacking means joining two or more arrays together.

**Syntax**
`np.vstack((a,b))` → stack vertically (row-wise)  
`np.hstack((a,b))` → stack horizontally (column-wise)  
`np.stack((a,b), axis=0)` → stack along new axis  

**Explanation**
`vstack` → arrays placed one below another  
`hstack` → arrays placed side by side  
`stack`  → creates new dimension  

**Note**
- Arrays should have same shape  
- Returns new array  
- Original arrays are not changed

```python
# horizontal stacking
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
a5
```

```python
np.hstack((a4,a5))
```

```python
# Vertical stacking
np.vstack((a4,a5))
```

### Splitting

### NumPy Splitting

**Definition:**  
Splitting means dividing an array into multiple smaller arrays.

**Syntax**
`np.split(a, sections)`      → split into equal parts  
`np.hsplit(a, sections)`     → split horizontally (column-wise)  
`np.vsplit(a, sections)`     → split vertically (row-wise)  
`np.array_split(a, sections)` → split (equal or unequal)

**Explanation**
`split`        → divides array into equal parts  
`hsplit`       → split by columns  
`vsplit`       → split by rows  
`array_split`  → allows unequal splits  

**Note**
- Returns multiple arrays  
- Original array not changed  
- Used to divide dataset
