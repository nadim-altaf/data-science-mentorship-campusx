# 4. Keywords & Identifiers

```python
# Keywords in Python:

# Predefined reserved words in Python that have special meaning and cannot be used as identifiers.

# Example: if, else, while, for, class, def, True, False, None.
```

```python
# Identifiers in Python:

# Names given to variables, functions, classes, etc., defined by the programmer. They must follow naming rules (cannot start with a digit, no special characters except _, case-sensitive).

# Example: age, student_name, calculateSum.
```

```python
# Rules for Identifiers in Python (short and standard):

# Only letters, digits, and underscore (_) are allowed.
# ✔ Example: name1, student_age, my_var

# Cannot start with a digit.
# ❌ 1value (invalid)
# ✔ value1 (valid)

# Cannot use keywords as identifiers.
# ❌ if, class, while

# Case-sensitive.
# Age and age are different identifiers.

# No special characters or spaces. we can use undrscore _
# ❌ my-name, student name

# Any length is allowed.
# But meaningful short names are recommended.
```

```python
# Arithmetric Operators
print(5+6) # Addition

print(5-6) # suntrction

print(5*6) # Mulplication

print(5/2) # division

print(5//2) # Integer devision (normal division convert karta hai interger mai)

print(5%2)   # Modulus -> gives remainder

print(5**2) # Power off
```

```python
# Guessing Game
import random

number = int(input('Guess the number'))

target = random.randint(1,100)
counter = 1
while number != target:
    if number < target:
        print('number is higher')
    else:
        print('number is lower')

    number = int(input('Guess the number'))
    counter += 1

else:
    print('correct guess')
    print('attempts',counter)

```

```python
# Write a program which can remove a particular character from a string.

s = input('enter a string')
term = input('what would like to remove ')

result = ''

for i in s:
    if i != term:
        result = result + i
print(result)

```

```python
# Write a program that can check whether a given string is palindrome or not.
# abba
# malayalam

s = input('enter the string')
flag = True
for i in range(0,len(s)//2):
  if s[i] != s[len(s) - i -1]:
    flag = False
    print('Not a Palindrome')
    break

if flag:
  print('Palindrome')

```

```python
# Write a program to count the number of words in a string without split()

s = input('enter the string')

L = []
temp = ''

for i in s:
    if i != ' ':
        temp  = temp + i
    else:
        L.append(temp)
        temp = ''
L.append(temp)
print(L)
```

```python
# Write a python program to convert a string to title case without using the title()

s = input('enter the string')
L = []
for i in s.split():
    L.append(i[0].upper() + i[1:].lower())
print(L)
```
