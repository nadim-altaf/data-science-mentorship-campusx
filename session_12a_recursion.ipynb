## Recursion using Python

```python
def multiply(a,b):
    result = 0

    for i in range(b):
        result = result + a

    print(result)

multiply(2,3)
```

```python
def mul(a,b):
    if b == 1:
        return a
    else:
        return a + mul(a,b-1)
    
print(mul(3,2))
```

```python
def fact(number):
    if number == 1:
        return 1
    else:
        return number * fact(number - 1)
print(fact(4))
  
    
```

```python
def palin(text):
    if len(text) <= 1:
        print( 'its Palindrome')
    else:
        if text[0] == text[-1]:
            palin(text[1:-1])
        else:
            print('not palindrome')

palin('madam')
palin('malayalam')
palin('python')

        
```
