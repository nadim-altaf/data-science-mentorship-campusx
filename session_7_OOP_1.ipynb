```python
class Atm:

    # constructor 
    def __init__(self):
        print(id(self))
        self.pin = ''
        self.balance = 0 
        # self.menu()
        
    
    def menu(self):
        user_input = input("""
        Hi How can I help you!
        1. Press 1 t create pin
        2. Press 2 to create pin
        3. Press 3 to chech balance
        4. Press 4 to withdraw
        5. Anything else to exit
        """)

        if user_input == '1':
            # create pin
            self.create_pin()
        elif user_input == '2':
            # change pin
            self.change_pin()
        elif user_input == '3':
            self.check_balance() 
            pass
        elif user_input == '4':
            self.withdraw()
            pass
        else:
            exit()
    
    def create_pin(self):
        user_pin = input('enter your pin')
        self.pin = user_pin

        user_balance = int(input('enter balance'))
        self.balance  = user_balance
        
        print('pin created sucessfully')
        self.menu()

    def change_pin(self):
        old_pin =input('enter old pin')
        if old_pin == self.pin:
            # let him change the pin 
            new_pin = input('enter new pin')
            self.pin = new_pin
            print('pin changed sucessful')
            self.menu()
        else:
            print('nahi kar sakhta hai change pin')
            self.menu()

    def check_balance(self):
        user_pin = input('enter your pin')
        if user_pin == self.pin:
            print('your balance is',self.balance)
        else:
            print('chal nikal yaha se')

    def withdraw(self):
        user_pin = input('enter your pin')

        if user_pin == self.pin:
            # allow user to withdraw
            amount = int(input('enter the amount'))
            if amount <= self.balance:
                self.balance = self.balance - amount
                print('withdraw is successful--balance is',self.balance)
            else:
                print('out of balance')

        else:
            print('wapas chale jav')
        self.menu()




```

#### Fraction Class

```python
class Fraction:
    def __init__(self,x,y): # parameterized constructor takes input as argument
        self.num = x
        self.den = y
    
    def __str__(self):
        return '{}/{}'.format(self.num,self.den)
    
    def __add__(self, other):
        new_num = self.num*other.den + other.num*self.den
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)
    
    def __sub__(self, other):
        new_num = self.num*other.den - other.num*self.den
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)
    
    def __mul__(self, other):
        new_num = self.num*other.num
        new_den = self.den*other.den

        return '{}/{}'.format(new_num,new_den)
    
    def __truediv__(self, other):
        new_num = self.num*other.den
        new_den = self.den*other.num

        return '{}/{}'.format(new_num,new_den)
    
    def convert_to_decimal(self):
        return self.num/self.den
    
```

```python
fr1 = Fraction(3,4)
fr2 = Fraction(1,2)
```

```python
fr1.convert_to_decimal()
```

```python
print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)
```
