## Session 6 - Functions in Python

`Function` (Python):<br>
A function is a reusable, well-defined block of code that takes input (parameters), performs a specific computation or transformation, and returns an output.
In Data Science, functions are used to clean data, transform features, apply statistical operations, build pipelines, and ensure reproducibility, readability, and efficiency in analysis and models.

`Why functions are essential in Data Science`

- Reusability → write once, use on many datasets
- Modularity → break complex workflows (EDA, preprocessing, modeling) into steps
- Reproducibility → same function = same result every time
- Clean pipelines → easy integration with ML workflows
- Debugging & testing → isolate logic

#### `*args and **kwargs`
`*args` and `**kwargs` are special Python keywords that are used to pass the variable length of arguments to a function

```python
# *args

# allows us to pass a variable number of non-keyword arguments to a function.

def multiply(*args):
    product = 1

    for i in args:
        product = product * i
    print(args)
    return product
```

```python
multiply(1,2,3,4,5,6)
# practical knowledge gives job, not only theory
```

```python
# **kwargs
# **kwargs allows us to pass any number of keyword arguments.
# Keyword arguments mean that they contain a key-value pair, like a Python dictionary.

def display(**kwargs):
    for (key,value) in kwargs.items():
        print(key,':',value)
```

```python
display(name='nadim',name1='aqib',name2='amaan',name3='ayaan')
```

Points to remember while using *args and **kwargs
- `order of the arguments matter(normal -> *args -> **kwargs)`
- The words “args” and “kwargs” are only a convention, you can use any name of your choice

Python uses pass by object reference
(also called pass by assignment).
Python has neither pure pass by value nor pass by reference; it uses pass by object reference.

```python
def f(x):
   x = x + 1
   print('in f(x): x =', x)
   return x

x = 3
z = f(x)
print('in main program scope: z =', z)
print('in main program scope: x =', x)
```

#### Lambda Function
A lambda function is a small anonymous function.

A lambda function can take any number of arguments, but can only have one expression.
![alt text](download-3.png)

```python
# x -> x^2
a = lambda x:x**2
a(2)
```

```python
# x,y -> x+y 
a = lambda x,y: x+y
a(2,3)
```

#### Diff between lambda vs Normal Function
- No name
- lambda has no return value(infact,returns a function)
- lambda is written in 1 line
- not reusable

Then why use lambda functions?

They are used with HOF(Higher Order Function)

```python
# check if a string has 'a' 
a = lambda s: 'a' in s
a('hello')
```

```python
# odd or even 
c= lambda a: 'even' if a%2==0 else 'odd'
c(9)
```

#### Higher order function
A function that takes another function as an argument or returns a function.

In one line:<br>
Function that works with other functions.

```python
# example
# def square(x):
#     return x**2

# def cube(x):
#     return x**3


# HOF
def transform(f,L):
    output = []
    for i in L:
        output.append(f(i))
    print(output)
L = [1,2,3,4,5]

# transform(square,L)
transform(lambda x:x*2,L)
```

#### Map

```python
# square the items of a list
L = [1,2,3,4,5]
list(map(lambda x:x**2,L))
```

```python
# odd/even labelling of list items
L = [1,2,3,4,5]
list(map(lambda x:'even' if x%2==0 else "odd",L))
```

```python
# fetch names from a list of dict
users = [
    {
        'name':'Rahul',
        'age':45,
        'gender':'male'
    },
    {
        'name':'Nitish',
        'age':33,
        'gender':'male'
    },
    {
        'name':'Ankita',
        'age':50,
        'gender':'female'
    }
]

list(map(lambda users:users['name'],users))
```

#### Filter

```python
# numbers greater than 5
L = [1,2,3,4,5,6,7,8,9]
list(filter(lambda x:x>5,L))
```

```python
# fetch fruits starting with 'a'
fruits = ['apple','guava','cherry']

list(filter(lambda x:x.startswith('a'),fruits))
```
