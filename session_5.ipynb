#### Tuples
A tuple in Python is similar to a list. The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.

Characterstics

- Ordered
- Unchangeble
- Allows duplicate

#### Plan of attack

- Operations on Tuples
- Tuple Functions

#### Creating Tuples

```python
# empty
t1 = ()
print(t1)

# create a tuple with a single item
t2 = (2,) # for single item/element we need to write ','/comma at the end to know it is tuple
print(type(t2))

# homo
t3 = (1,2,3,4,5)
print(t3)

# hetro
t4 = (1,2.4,True,[1,2,3])
print(t4)

# tuple
t5 = (1,2,3,4,(5,6,7)) # tuple inside tuple
print(t5)

# using type conversion
t6 = tuple("hello")
print(t6)
```

#### Accessing Items

#### Difference between Lists and Tuples
- Syntax -> list[ ] ---tuple()
- Mutability -> list are mutable and tuples are immutable
- Speed -> tuples are faster then list, generally immutable datatyple are faster
- Memory -> tuple take less memory then list, generally immutable datatyple takes less memory then mutable datatypes
- Built in functionality -> have more in list in camparision to tuple
- Error prone -> in list have more diffculty in debugging then tuples
- Usability -> for read only application we use tuple where as user need to do changes we prefer List 

```python
import time 

L = list(range(100000000))
T = tuple(range(100000000))

start = time.time()
for i in L:
  i*5
print('List time',time.time()-start)

start = time.time()
for i in T:
  i*5
print('Tuple time',time.time()-start)
```

```python
import sys

L = list(range(1000))
T = tuple(range(1000))

print('List size',sys.getsizeof(L))
print('Tuple size',sys.getsizeof(T))
```

#### Sets
A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:

- Unordered
- Mutable
- No Duplicates
- Can't contain mutable data types

```python
# empty
s = {} # wrong syntex for empty set, it creates dictionary
s1 = set()
print(s)
print(type(s))
print(s1)
print(type(s1))

# 1D and 2D
s2 = {1,2,3}
print(s2)

# 2D set is not possible, you cannoy=t have mutable datatype inside the set
# s3 = {1,2,3,{4,5,6}}

# homo and hetro
s4 = {1,2.3,'hello',True,(1,2,3)} # python treats True as a 1 , so 1 is already in set so no duplicates
print(s4)
# output is -> {'hello', 1, 2.3, (1, 2, 3)} -- order is diffrent because hashing decide the position of item

# using type conversion

s5 = set([1,2,3,4]) # ---> list to set
print(s5)

# duplicates not allowed
s6 = {1,2,2,3,4,5,5,5,6}
print(s6)

# set can't have mutable items

s7 = {1,2,[4,5]}
print(s7) # --> gives error 
```

```python
s1 = {1,2,3}
s2 = {3,2,1}

print(s1 == s2)
```

#### Set Operaton

```python
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# Uniom(|)
s1 | s2 # Combines all unique elements from both sets

# Intersection(&)
s1 & s2 # Returns only common elements present in both sets

# Difference(-)
s1 - s2 # Elements present in s1 but not in s2
s2 - s1 # Elements present in s2 but not in s1

# Stmmetric Difference(^)
s1 ^ s2 # Elements present in either set, but not in both

# Union → all
# Intersection → common
# Difference → left only
# Symmetric Difference → uncommon

# Membership Test

1 in s1

1 not in s2

# Iteration

for i in s1:
    print(i)
```

#### Set Function

```python
# len/ sum/ min/ max/ sorted
s = {5,3,2,4,1}

print(len(s)) # len
print(sum(s)) # sum
print(min(s)) # min
print(max(s)) # max
print(sorted(s)) # sorts the items, result od sorted is alqays a list
print(sorted(s,reverse=True))
```

```python
# union/update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

# s1 | s2
s1.union(s2)

s1.update(s2) # result of union s1 and s2 is stored in s1
print(s1)
print(s2) # s2 remains as it is
```

```python
# intersection/intersection_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.intersection(s2) 

s1.intersection_update(s2) # result of intersection s1 and s2 is stored in s1
print(s1)
print(s2)
```

```python
# difference/difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.difference(s2) 

s1.difference_update(s2) # result of difference s1 and s2 is stored in s1
print(s1)
print(s2)
```

```python
# symmetric_difference/symmetric_difference_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}

s1.symmetric_difference(s2)

s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint/issubset/issuperset
s1 = {1,2,3,4}
s2 = {4,5,6,7}

s1.isdisjoint(s2)
```

```python
# issubset
s1 = {1,2,3,4,5}
s2 = {3,4,5}

s1.issubset(s2)
s2.issubset(s1)
```

```python
# issuperset
s1 = {1,2,3,4,5}
s2 = {3,4,5}

s1.issuperset(s2)
```

```python
# copy
s1 = {1,2,3}

s2 = s1.copy()
print(s1)
print(id(s1))
print(s2)
print(id(s2))
```

#### Dictionary
Dictionary in Python is a collection of keys values, used to store data values like a map, which, unlike other data types which hold only a single value as an element.

In some languages it is known as map or assosiative arrays.

dict = { 'name' : 'nitish' , 'age' : 33 , 'gender' : 'male' }

Characterstics:

- Mutable
- Indexing has no meaning
- keys can't be duplicated -> Keys are unique and immutable (int, str, tuple)
- keys can't be mutable items -> like List, Sets , dictionary
- Values can be any data type

#### Why Dictionary is Important in Data Science 📊

- Represents structured data (JSON, API responses)<br>
- Used in feature mapping<br>
- Counting frequencies (EDA)<br>
- Label encoding (basic)<br>
- Intermediate data storage
