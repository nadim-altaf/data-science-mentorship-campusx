Buffer Memory in File Handling 

Buffer memory is a temporary memory area in RAM used to store data before it is written to or read from a file.

It improves file operation speed.

`How it Works`<br
When writing to a file:

1. Data first goes to buffer memory (RAM).
2. Then it is transferred to the file on disk.

`Program → Buffer (RAM) → Disk File`

When reading a file:

Disk File → Buffer (RAM) → Program

in buffer file is read character by character

#### How open() Works

1. Python searches for the file in the current working directory (or given path).

2. It checks the mode (r, w, a, etc.).

3. Python asks the operating system to open the file.

4. A file object is created in RAM, which allows operations like read() and write().

5. The actual file is stored on disk (HDD/SSD).

RAM → file object (f)<br>
Disk → actual file (file.txt)

`Important`

- Files are stored in the current working directory by default.

You can check it using:

import os<br>
print(os.getcwd())

### Using Context Manager (With)

- It's a good idea to close a file after usage as it will free up the resources
- If we dont close it, garbage collector would close it
- with keyword closes the file as soon as the usage is over

```python
# # with 
with open('sample1.txt','w') as f:
    f.write('hello')
```

#### seek and tell function

Function ------ Purpose

- tell() ------ Returns current file pointer position
- seek() ------ Moves file pointer to a specified position

```python
L = ['hi\n','hello\n','hi how are you','im fine']
with open('sample2.txt','w') as f:
    f.writelines(L)
```

```python
# seek and tell function

with open('sample2.txt','r') as f:
    f.seek(15)
    print(f.read(10))
    print(f.tell())

    print(f.read(10))
    print(f.tell())

```

### Serialization and Deserialization

- **Serialization** - process of converting python data types to JSON format
- **Deserialization** - process of converting JSON to python data types

#### What is JSON?
it is a universal data/text, format understood by all the programming language <br>
JSON data types include String, Number, Object, Array, Boolean, and Null.
![alt text](download-4.png)

```python
# serialization using json module
# list
import json
L = [1,2,3,4]

with open('demo.json','w') as f:
    json.dump(L,f)
```

```python
# dict 
# dict
d = {
    'name':'nitish',
     'age':33,
     'gender':'male'
}

with open('demo2.json','w') as f:
    json.dump(d,f,indent=4)
```

```python
# deserialization

import json

with open('demo2.json','r') as f:
    d = json.load(f)
    print(d)
    print(type(d))
```

```python
# serialize and deserialize tuple
import json

t = (1,2,3,4,5)

with open('demo3.json','w') as f:
    json.dump(t,f) # stores tuple as a list
```

```python
# serialize and deserialize a nested dict

d = {
    'student':'nitish',
     'marks':[23,14,34,45,56]
}

with open('demo.json','w') as f:
  json.dump(d,f)
```

### Serializing and Deserializing custom objects

```python
class Person:

  def __init__(self,fname,lname,age,gender):
    self.fname = fname
    self.lname = lname
    self.age = age
    self.gender = gender

# format to printed in
# -> Nadim Altaf age -> 23 gender -> male

```

```python
person = Person('Nadim','Altaf',23,'male')
```

```python
import json

def show_object(person):
    if isinstance(person,Person):
        return "{} {} age -> {} gender -> {}".format(person.fname,person.lname,person.age,person.gender)
with open('demo4.json','w') as f:
    json.dump(person,f,default=show_object,indent=4)
```

```python
# As a Dictionary 
import json

def show_object(person):
    if isinstance(person,Person):
        return {'name': person.fname + ''+ person.lname,'age':person.age, 'gender':person.gender}
with open('demo4.json','w') as f:
    json.dump(person,f,default=show_object,indent=4)  # indent arrtribute As a dict
```

```python
# deserialization 

import json

with open('demo4.json','r') as f:
    d = json.load(f)
    print(d)
    print(type(d))
```

### Pickling
`Pickling` is the process whereby a Python object hierarchy is converted into a byte stream, and `unpickling` is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy.

```python
class Person:

  def __init__(self,name,age):
    self.name = name
    self.age = age

  def display_info(self):
    print('Hi my name is',self.name,'and I am ',self.age,'years old')
```

```python
p = Person('nitish',33)
```

```python
# pickle dump
import pickle
with open('person.pkl','wb') as f:
  pickle.dump(p,f)
```

```python
# pickle load
import pickle
with open('person.pkl','rb') as f:
  p = pickle.load(f)

p.display_info()
```

### Pickle Vs Json

- Pickle lets the user to store data in binary format. JSON lets the user store data in a human-readable text format.
