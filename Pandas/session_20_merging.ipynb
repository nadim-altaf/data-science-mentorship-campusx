# Summary — Merging & Concatenation (expanded notes)

Below are short explanations and tiny examples for common merging/combining patterns in pandas:

- **pd.concat(list, axis=0)** — stack rows from multiple DataFrames. Use `ignore_index=True` to reindex the result.
  ```python
  pd.concat([df1, df2], ignore_index=True)
  ```

- **pd.concat(list, axis=1)** — join DataFrames column-wise (aligns on index). Useful when you have matching row order.
  ```python
  pd.concat([left_df, right_df], axis=1)
  ```

- **keys=...** with `concat` creates a MultiIndex (useful for labeling pieces). Flatten with `.reset_index()` if needed.
  ```python
  combined = pd.concat([nov, dec], keys=['Nov','Dec']).reset_index()
  ```

- **pd.merge / df.merge** — relational joins. Main params: `how` and `on`/`left_on`/`right_on`.
  ```python
  pd.merge(left, right, how='inner', on='key')
  pd.merge(left, right, how='left', left_on='id_left', right_on='id_right')
  ```

- **how** options: `inner` (intersection), `left` (all left rows), `right` (all right rows), `outer` (union; missing values become NaN).

- **Chained merges**: merge sequentially to combine >2 tables (e.g., regs -> students -> courses).
  ```python
  regs.merge(students, on='student_id').merge(courses, on='course_id')
  ```

- **groupby + agg**: aggregate after merge to compute totals or counts (revenue per course, enrollments per student).
  ```python
  regs.merge(courses, on='course_id').groupby('course_name')['price'].sum()
  ```

- **Membership & set ops**: use `np.intersect1d` / `np.setdiff1d` and `Series.isin()` to find common/different IDs and filter rows.
  ```python
  common = np.intersect1d(nov['student_id'], dec['student_id'])
  students[students['student_id'].isin(common)]
  ```

- **Counting / frequencies**: `value_counts()` or `groupby().count()` for quick frequency stats (e.g., matches per venue).
  ```python
  matches['venue'].value_counts()
  ```

- **drop_duplicates(subset=...)**: keep one row per key after sorting (useful to pick season winners after sorting by score).

Tips: prefer `pd.concat` over `DataFrame.append()` (deprecated); always choose `how` explicitly for merges; inspect NaNs after outer joins and use `fillna()` if appropriate.
