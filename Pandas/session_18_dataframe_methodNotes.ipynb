# 🐼 Pandas DataFrame Methods

| Method | Description | Example |
|---|---|---|
| `value_counts()` | Counts how many times each unique value appears in a column and returns them sorted from most to least frequent | `df['col'].value_counts()` |
| `sort_values()` | Sorts the rows of a DataFrame based on the values in one or more columns, either ascending or descending | `df.sort_values(by='age', ascending=False)` |
| `rank()` | Assigns a rank number to each value in a column — smallest gets rank 1. You can control how ties are handled using `method=` | `df['col'].rank(method='dense')` |
| `sort_index()` | Sorts the DataFrame by its index labels (not the data values). Useful when the index is scrambled after merging | `df.sort_index(ascending=False)` |
| `set_index()` | Replaces the default integer index with values from an existing column | `df.set_index('id')` |
| `rename()` | Renames column headers or index labels by passing a dictionary of `{old_name: new_name}` | `df.rename(columns={'old': 'new'})` |
| `reset_index()` | Moves the current index back into a regular column and replaces it with a default 0, 1, 2... integer index | `df.reset_index(drop=True)` |
| `unique()` | Returns an array of all distinct values in a Series, in the order they first appear | `df['col'].unique()` |
| `nunique()` | Returns the number of distinct values in a Series or DataFrame as a single integer | `df['col'].nunique()` |
| `isnull()` | Returns a DataFrame of True/False values — True wherever a value is missing (NaN) | `df.isnull().sum()` |
| `notnull()` | Opposite of `isnull()` — returns True wherever a value exists (is not NaN) | `df['col'].notnull()` |
| `hasnans` | A quick True/False property that tells you whether a Series contains at least one missing value | `df['col'].hasnans` |
| `dropna()` | Removes rows (or columns) that contain missing values. You can control how strict it is using `how=` and `thresh=` | `df.dropna(subset=['col'], how='any')` |
| `fillna()` | Replaces missing values with a specified value, a per-column dictionary, or by carrying forward/backward neighboring values | `df.fillna({'age': 0, 'name': 'N/A'})` |
| `drop_duplicates()` | Removes rows where all (or a subset of) columns have the same values as a previous row. You choose which copy to keep | `df.drop_duplicates(subset=['email'], keep='last')` |
| `drop()` | Deletes one or more rows or columns from the DataFrame by their label names | `df.drop(['col1', 'col2'], axis=1)` |
| `apply()` | Applies a custom function to each row or each column. Use it when built-in methods are not flexible enough | `df.apply(lambda row: row['a'] + row['b'], axis=1)` |
| `isin()` | Checks whether each value in a column belongs to a given list. Returns a True/False mask useful for filtering rows | `df[df['city'].isin(['Paris', 'London'])]` |
| `corr()` | Calculates the correlation between every pair of numeric columns. Result is a matrix with values between -1 and 1 | `df.corr(method='spearman')` |
| `nlargest()` | Returns the top N rows with the highest values in a specified column. Faster than sorting the whole DataFrame | `df.nlargest(5, 'sales')` |
| `nsmallest()` | Returns the top N rows with the lowest values in a specified column. Faster than sorting the whole DataFrame | `df.nsmallest(3, 'price')` |
| `insert()` | Adds a new column at a specific position in the DataFrame. Unlike `df['col'] =`, which always adds to the end | `df.insert(loc=0, column='id', value=range(len(df)))` |
| `copy()` | Creates a completely independent copy of the DataFrame so that changes to the copy do not affect the original | `df2 = df[df['age'] > 18].copy()` |
